## Step 1. Create two different dataframes in two different ways in the following cells. Dataframes should have at least five rows and two columns. 

Hint: If you need a reminder, go over module 5.4 or Google examples.

```python
#Need sample dictionary and sample np.array
a = {'col1':[2,3,5,6,7],'col2':[4,3,12,7,8]}
b = np.array([[1,2],[4,2],[3,6],[6,4],[5,6]])
cols = ['col1','col2']
```

```python
#create a dataframe called df1 using pd.DataFrame(dictionary)
#Replace ... with your code.
df1 = pd.DataFrame(a)
df1
```

```python
#create a dataframe called df2 using pd.DataFrame(np.array,columns=cols)
df2 = pd.DataFrame(b, columns=cols)
df2
```

```python
df1.loc[2, 'col2']
```
